```python
best_score = -math.inf
best_node = None

for goal, node in nodes.items():
    if node.is_explored or node.ignore:
        continue
    # take the score for a node as the probability of proving that goal,
    # multiplied by the probability of proving the best context of that goal
    # (i.e how likely to prove the original goal, assuming this goal is used)
    if node.context and len(node.context[0]) > 0:
        score = node.provable_score + max(
            [sum([nodes[ctx].up_score for ctx in context]) for context in node.context])
    else:
        score = node.provable_score

    if score > best_score:
        best_score = score
        best_node = node
```

```python
# check all ignored nodes have a proven ancestor
from updown_step.search_tree_updown import Status
def check_ignored(node):
    if node.ignore:
        assert any([nodes[a].status == Status.PROVED for a in node.ancestors])
    if not node.is_explored:
        if any([nodes[a].status == Status.PROVED for a in node.ancestors]):
            assert node.ignore

```

```python
def check_ancestors(node, ancestors):
    for parent in node.in_edges:
        parent = parent.src
        assert parent.goal in ancestors, (ancestors, node, parent)
        check_ancestors(parent, ancestors - {parent.goal})
```

```python
for node in nodes.values():
    check_ancestors(node, node.ancestors)
    check_ignored(node)
```
